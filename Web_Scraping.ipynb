```python
import requests
from bs4 import BeautifulSoup

# Target URL
url = "http://books.toscrape.com/catalogue/category/books/science_22/index.html"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}

# Send request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Convert rating text to numeric value
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

# Find all books on the page
books = soup.find_all("article", class_="product_pod")

# Extract book title and rating
for book in books:
    title = book.h3.a["title"]
    rating_class = book.p["class"][1]  # Example: "Three"
    rating = rating_map.get(rating_class, "N/A")  # Convert to number

    print(f"📖 {title} - ⭐ {rating} Stars")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
url="https://www.themoviedb.org/tv"
r=requests.get(url)
soup=BeautifulSoup(r.content,"lxml")





```

### Creating dataset from different tabs

```python


scrap=["tv","person","movie"]

for i in scrap:
    url="https://www.themoviedb.org/{}".format(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"lxml")
    if(i=="tv"):
        print("Scrapping data from tv tab...............\n")

        name=[]
        date=[]

        for i in soup.find_all("h2") :
          if(i.a!=None):
            name.append(i.a["title"])

        for i in soup.find_all("div", class_="card style_1"):
          j = i.find("p")  # Find the <p> inside the card
          if j:
            date.append(j.text.strip())

        tv=pd.DataFrame({"Name": name, "Date":date})


        print("Scrapped Successfully\n")

    if(i=="person"):
        print("Scrapping data from person tab...............\n")
        name=[]
        movie_name=[]

        for i in soup.find_all("p",class_='name'):
            name.append(i.text.strip())

        for i in soup.find_all("p",class_='sub'):
            movie_name.append(i.text.strip())

        person=pd.DataFrame({"name":name, "movie name": movie_name})
        print("Scrapped Successfully\n")


    if(i=="movie"):
      print("Scrapping data from movie tab...............\n")

      name=[]
      date=[]

      for i in soup.find_all("h2") :
        if(i.a!=None):
          name.append(i.a["title"])

      for i in soup.find_all("div", class_="card style_1"):
        j = i.find("p")  # Find the <p> inside the card
        if j:
          date.append(j.text.strip())

      movie=pd.DataFrame({"Name": name, "Date":date})



      print("Scrapped Successfully\n")




```
